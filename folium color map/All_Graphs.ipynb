# Creating a Population Density Map with Folium

In the next section, We will use Folium to create a map, and zoom into a specific section.
Then, we need to read out GeoJson file to determine boundaries for each county, and then read in all the data using pandas.
Using Folium, we are able to parse that geojson file and tie the CSV data to each specific county.

The data is saved into "popData.json", and the map is saved to "pop.html"

For specifics on the geo_json method, we first speccify the geoJson file location, the excel sheet data, and then the data output. The columns are the columns in the excel sheet used to produce the data. The key_on is what ties each county boundary to a specific id of data. The fill color is the map color scheme, and the opacity is how dark each map is. The legend name determines the name of the legend in the top right of each map.

```python

map_osm_pop = folium.Map(location = [37.681229,-117.258796691895],zoom_start=6)

ca_counties = r'california-counties.geojson'

df = pd.read_csv(r'county_full.csv')

map_osm_pop.geo_json(geo_path=ca_counties, data=df, data_out = 'popData.json',
             columns=['id','popu_density'],
             key_on='feature.properties.cartodb_id',
             fill_color='BuPu', fill_opacity=0.4, line_opacity=0.3,
             legend_name='Population Density (Person / Sq. Mile)')


map_osm_pop.create_map('pop.html')

map_osm_pop
```

# Water Quality Graphs

The graphs shown below are all very similar in how they were created, with the only major differences being which column we got the data from. Also, we changed the threshold scale on some of the graphs to more accurately reflect the data visually.

The maps are saved to differently named html files.

```python
map_osm_oxy = folium.Map(location = [37.681229,-117.258796691895],zoom_start=6)

map_osm_oxy.geo_json(geo_path=ca_counties, data=df, data_out = 'oxyData.json',
             threshold_scale = [5,7,9,11,13,15],
             columns=['id','median_oxy_Result'],
             key_on='feature.properties.cartodb_id',
             fill_color='BuPu', fill_opacity=0.4, line_opacity=0.3,
             legend_name='Median Oxygen Dissolved')


map_osm_oxy.create_map('oxygen.html')

map_osm_oxy
```
